```python
import numpy as np
import matplotlib.pyplot as plt
from math import log, exp
```

```python
def SoftPlusActivation(weight):
    return log(1 + exp(weight))
```

So These weights are determine by using backporopagation....
##First Node

```python

weights = np.arange(0, 1.0, 0.1)
y = 0

yVal = []

for weight in weights:
    x = weight * -34.4 + 2.14
    y = round(SoftPlusActivation(x), 2)
    yVal.append(y)

plt.scatter(np.arange(len(yVal)), yVal, color='blue', label='Y Values')

plt.xlabel('Index')
plt.ylabel('softplus y value')
plt.title('y value')
plt.legend()
plt.show()
```

##Second Node

```python

def softPlus(x):
    return log(1 + exp(x))

weights = np.arange(0, 1.0, 0.1)
y = 0

yVal = []

for weight in weights:
    x = weight * -2.52 + 1.29
    y = round(SoftPlusActivation(x), 2)
    yVal.append(y)

plt.scatter(np.arange(len(yVal)), yVal, color='blue', label='Y Values')
plt.xlabel('Index')
plt.ylabel('softplus y value')
plt.title('y value')
plt.legend()
plt.show()
```

#Adding The Second Bias

```python


yVal0 = []
yVal1 = []



def softPlus(x):
    return log(1 + exp(x))

weights = np.arange(0, 1.0, 0.1)
yVal0 = []
yVal1 = []

# First Node
for wieght in weights:
    x = wieght * -34.4 + 2.14
    y = round(softPlus(x), 2)
    yBias = y * -1.30
    yVal0.append(yBias)


# Second node
for wieght in weights:
    x = wieght * -2.52 + 1.29
    y = round(softPlus(x), 2)
    yBias = y * 2.28
    yVal1.append(yBias)



plt.scatter(np.arange(len(yVal0)), yVal0, color='blue', label='Y Values')

plt.scatter(np.arange(len(yVal1)), yVal1, color='blue', label='Y Values')
plt.xlabel('Index')
plt.ylabel('softplus y value')
plt.title('y value')
plt.legend()
plt.show()
```

#Sum it Up with All the val in yval0 and yval1

```python
yValSum = [round(y0 + y1 - 0.58, 2) for y0, y1 in zip(yVal0, yVal1)] # 0.58 Final Bias


# Plot the squiggly line using the yValSum values
plt.plot(weights, yValSum, color='blue', linestyle='-', marker='o', label='yValSum')

# Add labels and title
plt.xlabel('Dosage')
plt.ylabel('Efficency')
plt.title('Squiggly Line Plot for yValSum')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
